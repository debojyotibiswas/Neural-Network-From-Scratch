```python
from sklearn import datasets
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
print(X)
```

```python
X_ = []
for i in range(len(X)):
    X_.append(np.insert(X[i], 0 , 1))
X_ = np.array(X_)
X_
```

```python
y_ = iris.target
y_
```

```python
X_, y_ = shuffle(X_, y_)
```

```python
test_size = int(0.3 * len(X_))
X_train, X_test = X_[:test_size], X_[test_size:]
y_train, y_test = y_[:test_size], y_[test_size:]
```

```python
class MC_Perceptron:
    
    def __init__(self, eta = 0.01, n_iters = 500):
        self.eta = eta
        self.n_iters = n_iters
        self.weight = np.zeros((3,5))
        self.error = []
        
    def train(self, X, y):
        dim = X.shape[1]
        self.weight[0] = np.insert(np.zeros(dim-1), 0, 0)
        self.weight[1] = np.insert(np.zeros(dim-1), 0, 1)
        self.weight[2] = np.insert(np.zeros(dim-1), 0, 2)
        
        for _ in range(self.n_iters):
            
            m = 0
            for idx, x_i in enumerate(X):
                
                w = []
                w.append(self.weight[0].dot(x_i))
                w.append(self.weight[1].dot(x_i))
                w.append(self.weight[2].dot(x_i))
                
                y_pred = self.transfer_function(w)
                
                #learning rule
                if y_pred != y[idx]:
                    
                    m += (y_pred - y[idx])**2
                    self.weight[y[idx]] = self.weight[y[idx]] + self.eta * x_i
                    self.weight[y_pred] = self.weight[y_pred] - self.eta * x_i
            self.error.append(m)
            
    def predict(self, X):
        
        y_pred = []
        for i in range(len(X)):
            w = []
            w.append(self.weight[0].dot(X[i]))
            w.append(self.weight[1].dot(X[i]))
            w.append(self.weight[2].dot(X[i]))
            y_pred.append(self.transfer_function(w))
        return np.array(y_pred)
            
                
    def transfer_function(self, w):
        return np.argmax(np.array(w))
    
    def confusion_matrix(self, y, y_pred, y_actual):
        
        classes = np.unique(y_actual)
        
        conf_mat = np.zeros((len(classes), len(classes)))
        for i in range(len(classes)):
            for j in range(len(classes)):
                conf_mat[i, j] = np.sum((y == classes[i]) & (y_pred == classes[j]))
        return conf_mat
       
    def error_iteration_curve(self):
        
        fig = plt.figure(figsize=(20,9))
        plt.xlabel('Iteration no.')
        plt.ylabel('Error')
        plt.plot(self.error,color = 'tab:red')
        plt.show()
        
        
```

```python
p = MC_Perceptron()
p.train(X_train, y_train)
```

```python
y_pred = p.predict(X_test)
```

```python
conf = p.confusion_matrix(y_test,y_pred, y_)
conf 
```
